# Contour Detection – Label

# Detecting Product Shapes using Contours

```python
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
```

```python
contour_img = gray.copy()
```

```python
contour_img = cv2.cvtColor(contour_img, cv2.COLOR_GRAY2BGR)
```

Draw all contours (in green)

```python
cv2.drawContours(contour_img, contours, -1, (0, 255, 0), 2)
```

```python
plt.figure(figsize=(10,6))
```

```python
plt.imshow(cv2.cvtColor(contour_img, cv2.COLOR_BGR2RGB))
```

```python
plt.title("Step 3 – Detected Product Contours on Shelf")
```

# Step-4: Draw Bounding Boxes Around Contours

```python

output_img = gray.copy()
output_img = cv2.cvtColor(output_img, cv2.COLOR_GRAY2BGR)

for contour in contours:
    x, y, w, h = cv2.boundingRect(contour)
    cv2.rectangle(output_img, (x, y), (x + w, y + h), (0, 255, 0), 2)

plt.figure(figsize=(10, 6))
plt.imshow(cv2.cvtColor(output_img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Detected Objects with Bounding Boxes")
plt.show()
```

# how to count objects using OpenCV:

```python
print(f"Total Objects Detected: {len(contours)}")
```

# Filter tiny objects since to avoid counting tiny dots or noise as objects

```python
min_area = 500
valid_objects = []

for cnt in contours:
    area = cv2.contourArea(cnt)
    if area > min_area:
        valid_objects.append(cnt)

print(f"Filtered Object Count (Area > {min_area}): {len(valid_objects)}")
```

# Draw Bounding Boxes Around Detected Objects

```python

boxed_img = img.copy()
for cnt in valid_objects:
    x, y, w, h = cv2.boundingRect(cnt)
    cv2.rectangle(boxed_img, (x, y), (x + w, y + h), (255, 0, 0), 2)

plt.figure(figsize=(10, 6))
plt.imshow(cv2.cvtColor(boxed_img, cv2.COLOR_BGR2RGB))
plt.title("Detected Products with Bounding Boxes")
plt.axis("off")
plt.show()
```

# Count and Annotate Detected Products

```python

annotated_img = img.copy()

for i, cnt in enumerate(valid_objects):
    x, y, w, h = cv2.boundingRect(cnt)
    cv2.rectangle(annotated_img, (x, y), (x + w, y + h), (0, 255, 0), 2)
    cv2.putText(annotated_img, f"#{i+1}", (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX,
                0.6, (0, 0, 255), 2)

plt.figure(figsize=(10, 6))
plt.imshow(cv2.cvtColor(annotated_img, cv2.COLOR_BGR2RGB))
plt.title(f"Annotated Products: {len(valid_objects)} Detected")
plt.axis("off")
plt.show()
```

# Step_6:  Crop and Save Detected Products

```python
import os

output_dir = "cropped_products"
os.makedirs(output_dir, exist_ok=True)
for i, cnt in enumerate(valid_objects):
    x, y, w, h = cv2.boundingRect(cnt)
    cropped = img[y:y+h, x:x+w]
    filename = f"{output_dir}/product_{i+1}.jpg"
    cv2.imwrite(filename, cropped)
    print(f"Saved: {filename}")
```

# Step 7: Display All Cropped Product Images in a Grid

```python
import matplotlib.pyplot as plt
import cv2
import os
cropped_dir = "cropped_products"
cropped_files = sorted([f for f in os.listdir(cropped_dir) if f.endswith(".jpg")])

cols = 3
rows = (len(cropped_files) + cols - 1) // cols

plt.figure(figsize=(15, 5 * rows))

for i, file in enumerate(cropped_files):
    img_path = os.path.join(cropped_dir, file)
    img = cv2.imread(img_path)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.subplot(rows, cols, i + 1)
    plt.imshow(img_rgb)
    plt.title(f"Product {i+1}")
    plt.axis('off')

plt.tight_layout()
plt.show()
```

# Visualize Product Sales + Detected Products

```python
import matplotlib.pyplot as plt
from PIL import Image

detected_image = Image.open("/content/drive/MyDrive/AI_Business_visual_Dashboard/supermarket_front_photo.webp")
detected_image = detected_image.resize((400, 400))

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.imshow(detected_image)
plt.axis('off')
plt.title("Detected Products on Shelf")

plt.subplot(1, 2, 2)
product_summary.plot(kind='barh', color='skyblue')
plt.xlabel("Total Quantity Sold")
plt.title("Sales by Product Line")
plt.grid(axis='x')

plt.tight_layout()
plt.show()
```

# Product Sales by City (Bar Graph)

```python
city_sales = data.groupby('City')['Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(8, 5))
city_sales.plot(kind='bar', color='orange')
plt.title('Total Sales by City')
plt.xlabel('City')
plt.ylabel('Sales Amount')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Total Sales Per Day (Line Chart)

```python
data['Date'] = pd.to_datetime(data['Date'])

daily_sales = data.groupby('Date')['Sales'].sum()

plt.figure(figsize=(10, 5))
daily_sales.plot(kind='line', marker='o', color='green')
plt.title('Total Sales per Day')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Smart Alert: Peak Sales Detection

```python

threshold = daily_sales.mean() + daily_sales.std()

peak_days = daily_sales[daily_sales > threshold]

print(" Peak Sales Days Detected:")
print(peak_days)
```
